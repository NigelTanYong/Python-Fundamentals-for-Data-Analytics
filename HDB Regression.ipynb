```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
hdb_df = pd.read_csv("/content/drive/MyDrive/20240316 NTU Python Workshop/hdb_cleaned_data.csv")
hdb_df = pd.get_dummies(hdb_df, columns=["town"]) #make town items into one-hot encoding
# hdb_df = pd.get_dummies(hdb_df, columns=["town","flat_type"]) #make town items into one-hot encoding
hdb_df.head()
```

```python
train = hdb_df.sample(frac = 0.7, random_state=101) #sampling 70% of the dataset randomly. Random state locks in the 70% hosen at random at the start
test = hdb_df.drop(train.index) #test set drops all data points that are in the train set

trainY = train["resale_price"] #series
trainX = train[["remaining_lease", "floor_area_sqm"]] #dataframe ==>the factors that you think affects resale price

#Predicted Resale Price = (constant A)*Remaining Lease + (constant B)*Floor Area Sqm + constant c
#this training set will give you the line, and we will test it against the testing value

testY = test["resale_price"] #Actual resale price
testX = test[["remaining_lease", "floor_area_sqm"]]

model = LinearRegression().fit(trainX, trainY)
print(model.coef_)
print(model.intercept_)

#predicted Resale Price = 1301 * Remaining Lease + 3426 * Floor Area Sqm * 7769

predY = model.predict(testX)
mae = mean_absolute_error(testY, predY) #absolute differenece between the test and actual resale price
print(mae)
```
