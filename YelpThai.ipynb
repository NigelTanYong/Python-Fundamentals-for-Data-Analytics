```python
#importing dataset
import requests #library for python to process json and use in code => Requests is a simple, yet elegant, HTTP library.
re = requests.get("https://bit.ly/yelpthaifood") #get response request in json
thaifood = re.json() #converts the response from JSON format to a Python list
print(thaifood)
```

```python
#1: get all names of restaurants
namelist = []
for restaurant in thaifood:
  # print(restaurant["name"])
  namelist.append(restaurant["name"])
print(namelist)
```

```python
#2: get names of restaurants with rating above 3.5
restabove35=[]
for restaurant in thaifood:
  if restaurant["rating"] > 3.5:
    # print(restaurant["name"], "|rating:", restaurant["rating"])
    restabove35.append(restaurant["name"])
print(restabove35)
```

```python
#3: get names of restaurants with location distance below 2000
restbelow2k = []
for restaurant in thaifood:
  if restaurant["distance"] < 2000:
    # print(restaurant["name"], "|distance:", restaurant["distance"])
    restbelow2k.append(restaurant["name"])
print(restbelow2k)
```

```python
#4: BONUS get name of restaurants with rating above 3.5 and location distance below 2k
restabove35below2k = []
for restaurant in thaifood:
  if restaurant["rating"]> 3.5 and restaurant["distance"]< 2000:
    # print(restaurant["name"], "|ratings:", restaurant["rating"], "distance:", restaurant["distance"])
    restabove35below2k.append(restaurant["name"])
print(restabove35below2k)
```

```python
import pandas as pd #as is an alias
thai_df = pd.DataFrame(thaifood) #DataFrame
# thai_df.head() #print out first 5 items
thai_df.tail() #prints out last 5 items
```
