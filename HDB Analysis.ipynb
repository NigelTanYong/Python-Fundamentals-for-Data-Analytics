```python
#data visualisations

import plotly.express as px #alias

#scatterplot
##my hypotthesis: the longer the remaining lease, the higher the resale price

fig = px.scatter(hdb_df, x="remaining_lease", y="resale_price", hover_data=["town", "storey_range"])
fig.show()
#as remaining lease increases, the plot moving upwards
```

```python
#box plot
#hypothesis: for each town, we may need to predict the prices differently. TOWNSHIP AFFECTS RESALE PRICE

fig = px.box(hdb_df, x="town", y="resale_price")
fig.show()
#bigger box bigger variation ()
#smaller box smaller variation (better prediction as a factor for resale price analysis)
```

```python
#histogram/countplot
#hypothesis: higher supply --> lower prices

fig = px.histogram(hdb_df, x="town")
fig.show()
```

```python
#hypothesis - nearer to CBD, prices are more expensive

fig = px.scatter(hdb_df, x="lng", y="lat", color="resale_price")
fig.show()

#near to cbd area, more expensive
```

```python
#hypothesis - older lease_commence_date is cheaper
fig = px.scatter(hdb_df, x="lease_commence_date", y="resale_price", hover_data=["town", "storey_range"])
fig.show()
#trend increase slightly, newer more ex
```

```python
#hypothesis - flat model affects price
fig = px.box(hdb_df, x="flat_model", y="resale_price")
fig.show()
#in new gen, model a, simplified, improved, premium apartment & mansionette, there are lots of outlier. So cannot prove hypothesis
```
